## Model ćwiartkowy zawieszenia pojazdu kołowego
![](./Cwiartkowy_model.png)&nbsp; 

Na powyższym rysunku przedstawiony został model obliczeniowy ćwiartkowego, pasywnego układu zawieszenia pojazdu kołowego. Przyjęto oznaczenia:

- $m_s - $ masa wibroizolowana
- $m_t - $ masa niewibroizolowana
- $z_s - $ przemieszczenie masy wibroizolowanej
- $z_t - $ przemieszczenie masy niewibroizolowanej
- $z_w - $ wymuszenie
- $k_s - $ współczynnik sprężystości zawieszenia
- $k_t - $ współczynnik sprężystości opony
- $c_s - $ nieliniowy element tłumiący zawieszenia
- $c_t - $ współczynnik tłumienia opony

Równania równowagi sił dla każdej z mas:
$$
\begin{equation}
m_t\ddot{z}_t + F_c(\dot{z}_t - \dot{z}_s) + k_s(z_t - z_s) + c_t(\dot{z}_t - \dot{z}_w) + k_t(z_t - z_w) = 0 \\
m_s\ddot{z}_s + F_c(\dot{z}_s - \dot{z}_t) + k_s(z_s - z_t) = 0
\end{equation}
$$
<br></br>
Funkcja $F_c(v)$ opisuje nieliniową charakterystykę tłumika. Dana jest zależnością:

$$
\begin{equation}
F_c(v) = \left\{ \begin{array}{ll}
a_1v + b_1 & \textrm{gdy $v<v_1$}\\
a_2v + b_2 & \textrm{gdy $v \in \langle v_1,v_2)$}\\
a_3v + b_3 & \textrm{gdy $ v \geq v_2 $}
\end{array} \right.
\end{equation}
$$

## Odczyt pliku TDMS
Do odczytu pliku \*.tdms  używam biblioteki [nptdms](https://nptdms.readthedocs.io/en/latest/ "nptdms")

### Wyświetlanie parametrów pliku TDMS

```python
from nptdms import TdmsFile

file_path = "./Data/p2.tdms"
tdms_file = TdmsFile(file_path)
print("Reading TDMS file: " + file_path)
print("TDMS groups: ")
groups = tdms_file.groups()
print(groups, '\n')
for group in groups:
    print(group, '\n')
    for channel in tdms_file.group_channels(group):
        print(channel.channel, "[" + channel.property("unit_string") + "]")
```

### Odczyt danych do zmiennych

```python
file_path = "./Data/p2.tdms"
tdms_file = TdmsFile(file_path)
groups_list = tdms_file.groups()

disp_channel = tdms_file.object(groups_list[0], 'w')
force_channel = tdms_file.object(groups_list[0], 'SilaZEPWN_20kN')

displacement = disp_channel.data
displacement_unit = disp_channel.property("unit_string")
force = force_channel.data
force_unit = force_channel.property("unit_string")
time = disp_channel.time_track()
time_unit = disp_channel.property('wf_xunit_string')

recording_date = disp_channel.property('wf_start_time')
sampling_freq = int(1/(time[2]-time[1]))
```

### Odczyt parametrów pomiaru z nazwy grupy

```python
def read_group_properties(group_name):
    column1 = []
    column2 = []

    group_name_s1 = group_name.split('\t')
    topic = group_name_s1[0].split('=')[-1] #[-1] = last index

    group_name_s2 = group_name_s1[-1].split(';')

    for item in group_name_s2:
        item_splitted = item.split('=')
        column1.append(item_splitted[0])
        column2.append(item_splitted[-1])
        if "fqTAB" in item_splitted[0]:
            freq_string = item_splitted[1]
            freq_string_list = freq_string.split('|')
        if "Skladowa stala" in item_splitted[0]:
            shift_displacement = float(item_splitted[1])
        if "nT" in item_splitted[0]:
            nT = int(item_splitted[1])
        if "start phase" in item_splitted[0]:
            start_phase = float(item_splitted[1])
    return topic, column1, column2, freq_string_list, shift_displacement, nT, start_phase
            
group = groups_list[0]
topic, column1, column2, freq_string_list, shift_displacement, nT, start_phase = read_group_properties(group)
```

#### Konwersja listy string do float

```python
def convert_to_float(string_list):
    float_list = []
    for index, item in enumerate(string_list):
        try:
            item = float(item)
            float_list.append(item)
        except:
            print('Can not convert ' + str(index) + ' index ' + 'into float')       
    return float_list

freq_list = convert_to_float(freq_string_list)
```

## Analiza danych

### Eliminacja składowej stałej

```python
displacement_shifted = displacement - shift_displacement
```

### Wykresy przemieszczenia i siły od czasu

```python
import matplotlib.pyplot as plt
%matplotlib notebook

fig, axarr = plt.subplots(2, sharex=True, figsize=(9, 7))

axarr[0].plot(time, displacement_shifted, 'b', label = 'przemieszczenie')
axarr[0].set_title('Wykresy sygnałów')
axarr[0].legend()
axarr[0].grid()
axarr[0].set_ylabel('przemieszczenie [' + displacement_unit + ']')

axarr[1].plot(time, force, 'r', label = 'siła')
axarr[1].legend()
axarr[1].grid()
axarr[1].set_ylabel('siła [' + force_unit + ']')

plt.xlabel('czas [' + time_unit + ']')

fig.show()
```

### Wykresy przemieszczenia i prędkości od czasu

```python
import numpy as np

def smooth(y, box_pts):
    box = np.ones(box_pts)/box_pts
    y_smooth = np.convolve(y, box, mode='same')
    return y_smooth

velocity = np.diff(displacement_shifted)/(1/sampling_freq)
velocity_filtered = smooth(velocity, 21)
velocity_filtered = smooth(velocity_filtered, 7)

fig, axarr = plt.subplots(2, sharex=True, figsize=(9, 7))

axarr[0].plot(time, displacement_shifted, 'b', label = 'przemieszczenie')
axarr[0].set_title('Wykresy sygnałów')
axarr[0].legend()
axarr[0].grid()
axarr[0].set_ylabel('przemieszczenie [' + displacement_unit + ']')

axarr[1].plot(time[:-1], velocity, 'g,', label = 'prędkość')
axarr[1].plot(time[:-1], velocity_filtered, 'black', label = 'prędkość wygładzona')
axarr[1].legend()
axarr[1].grid()
axarr[1].set_ylabel('prędkość [' + displacement_unit + '/' + time_unit +']')

plt.xlabel('czas [' + time_unit + ']')

fig.show()
```

### Siła od przemieszczenia

#### Obliczenie okresu wymuszenia

```python
def calculate_period(freq_list, sampling_freq):
    period_list_samples = []
    for freq in freq_list:
        period = 1/freq
        period_samples = int(period * sampling_freq)
        period_list_samples.append(period_samples)
    return period_list_samples

period_list_samples = calculate_period(freq_list, sampling_freq)
```

#### Obliczenie przesunięcia fazowego

```python
def calculate_phase_shift(period_list_samples, start_phase):
    
    phase_shift_list = [] #samples
    phase = 360/start_phase
    for period in period_list_samples:
        phase_shift_samples = period/phase
        phase_shift_samples = phase_shift_samples/2
        phase_shift_list.append(int(phase_shift_samples))
    return phase_shift_list
        
phase_shift_list = calculate_phase_shift(period_list_samples, start_phase)
```

#### Wykres przemieszczenia i siły od czasu z zaznaczonymi przedziałami 

```python
start_signal = 838 #[samples]

fig, axarr = plt.subplots(2, sharex=True, figsize=(9, 7))

axarr[0].plot(time, displacement_shifted, 'b.', label = 'przemieszczenie')
axarr[0].set_ylabel('przemieszczenie [' + displacement_unit + ']')
axarr[0].plot(time[start_signal], displacement_shifted[start_signal], 'ro', label = 'start sygnału')

ss = start_signal
for i, period in enumerate(period_list_samples):
    for j in range(0, nT):
        start = ss + j * period - phase_shift_list[i]
        end = start + period
        if j == (nT - 1):
            axarr[0].axvspan(time[start], time[end], alpha=0.3, color='g')
            axarr[1].axvspan(time[start], time[end], alpha=0.3, color='g')
    ss = ss + period * nT
axarr[0].legend()
axarr[0].grid()

axarr[1].plot(time, force, 'r', label = 'siła')
axarr[1].set_ylabel('siła [' + force_unit + ']')
axarr[1].legend()
axarr[1].grid()

plt.xlabel('czas [' + time_unit + ']')
fig.tight_layout()
fig.show()
```

#### Utworzenie list dla przemieszczenia i siły

```python
def create_displacement_and_force_list(displacement_shifted, force, period_list_samples, phase_shift_list, start_signal, nT):
    displacement_list = []
    force_list = []
    ss = start_signal

    for i, period in enumerate(period_list_samples):
        for j in range(0, nT):
            start = ss + j * period - phase_shift_list[i]
            end = start + period
            if j == (nT - 1):
                displacement_cuted = displacement_shifted[start:end]
                displacement_list.append(displacement_cuted)
                force_cuted = force[start:end]
                force_list.append(force_cuted)
        ss = ss + period * nT
    return displacement_list, force_list

displacement_list, force_list = create_displacement_and_force_list(displacement_shifted, force, period_list_samples, phase_shift_list, start_signal, nT)
```

#### Wykres siły od przemieszczenia

```python
plt.figure(figsize=(9,7))

for i, disp in enumerate(displacement_list):
    #if 0<=i<12:
    plt.plot(disp, force_list[i], label = str(freq_list[int(i)]) + ' [Hz]')

plt.xlabel('przemieszczenie [' + displacement_unit + ']')
plt.ylabel('siła [' + force_unit + ']')
plt.legend()
plt.grid()
plt.show()
```

### Obliczenie pracy

Liczę pracę dla każdego cyklu używając, całkowania metodą trapezową

```python
def calculate_work(displacement_list, force_list):
    work_list = []
    for i, disp in enumerate(displacement_list):
        work = np.trapz(force_list[i], displacement_list[i])
        work_list.append(work/1000)
    return work_list
        
work_list = calculate_work(displacement_list, force_list)
```

#### Wykres pracy od częstotliwości wymuszczenia

```python
plt.figure(figsize=(5,5))
plt.title('Praca amortyzatora w całym zakresie przemieszczenia')
plt.plot(freq_list, work_list, 'b-o')
plt.xlabel('częstotliwość [Hz]')
plt.ylabel('praca [J]')
plt.grid()
plt.show()
```

#### Wykresy w czasie

```python
transient_time_coefficient = 0.25 #czas nieustalony 0.25
fixed_time_coefficient = 0.5 #czas ustalony 0.5
```

#### Utworzenie list dla odbicia i kompresji

```python
def create_rebound_and_compression_list(displacement_shifted, force, period_list_samples, phase_shift_list, start_signal, nT, transient_time_coefficient, fixed_time_coefficient):
    displacement_rebound_list = []
    displacement_compression_list = []
    force_rebound_list = []
    force_compression_list = []
    ss = start_signal

    for i, period in enumerate(period_list_samples):
        for j in range(0, nT):
            start = ss + j * period - phase_shift_list[i]
            end = start + period
            rebound_start = start + int(transient_time_coefficient*(period/2))
            rebound_end = rebound_start + int(fixed_time_coefficient*(period/2))
            compression_start = start + int(period/2) + int(transient_time_coefficient*(period/2))
            compression_end = compression_start + int(fixed_time_coefficient*(period/2))
            if j == (nT - 1):
                displacement_rebound_cuted = displacement_shifted[rebound_start:rebound_end]
                displacement_rebound_list.append(displacement_rebound_cuted)
                displacement_compression_cuted = displacement_shifted[compression_start:compression_end]
                displacement_compression_list.append(displacement_compression_cuted)
                force_rebound_cuted = force[rebound_start:rebound_end]
                force_rebound_list.append(force_rebound_cuted)
                force_compression_cuted = force[compression_start:compression_end]
                force_compression_list.append(force_compression_cuted)
        ss = ss + period * nT
    return displacement_rebound_list, displacement_compression_list, force_rebound_list, force_compression_list

transient_time_coefficient = 0.25 #czas nieustalony 0.25
fixed_time_coefficient = 0.5 #czas ustalony 0.5

displacement_rebound_list, displacement_compression_list, force_rebound_list, force_compression_list = create_rebound_and_compression_list(displacement_shifted, force, 
                                                                                                                                            period_list_samples, phase_shift_list, 
                                                                                                                                            start_signal, nT, 
                                                                                                                                            transient_time_coefficient, fixed_time_coefficient)
```

#### Obliczenie średniej wartości siły

```python
def calculate_mean(force_list):
    mean_force_list = []
    for force_array in force_list:
        mean_force = np.mean(force_array)
        mean_force_list.append(mean_force)
    return mean_force_list
    
rebound_mean_force_list = calculate_mean(force_rebound_list)
compression_mean_force_list = calculate_mean(force_compression_list)
```

#### Obliczenie tablicy prędkości
z częstotliwości wymuszenia

```python
def calculate_velocity(displacement, freq_list):
    distance_mm = round(max(displacement) - min(displacement))
    distance = distance_mm * 0.001
    compression_velocity_list = []
    rebound_velocity_list = []

    for freq in freq_list:
        compression_velocity = -2 * distance/(1/freq)
        rebound_velocity = 2 * distance/(1/freq)
        compression_velocity_list.append(round(compression_velocity, 2))
        rebound_velocity_list.append(round(rebound_velocity, 2))
    return rebound_velocity_list, compression_velocity_list

rebound_velocity_list, compression_velocity_list = calculate_velocity(displacement_shifted, freq_list)
```

##### Podzielenie wykresu na 3 części

```python
def divide_chart(all_velocity_list, velocity_1, velocity_2):
    part_1_velocity = []
    part_2_velocity = []
    part_3_velocity = []
    part_1_mean_force = []
    part_2_mean_force  = []
    part_3_mean_force  = []

    for index, velocity in enumerate(all_velocity_list):
        if velocity < velocity_1:
            part_1_velocity.append(velocity)
            part_1_mean_force.append(all_mean_force_list[index])
        if velocity_1 <= velocity < velocity_2:
            part_2_velocity.append(velocity)
            part_2_mean_force.append(all_mean_force_list[index])
        if velocity >= velocity_2:
            part_3_velocity.append(velocity)
            part_3_mean_force.append(all_mean_force_list[index])
    return part_1_velocity, part_2_velocity, part_3_velocity, part_1_mean_force, part_2_mean_force, part_3_mean_force

velocity_1 = -0.12
velocity_2 = 0.12

part_1_velocity, part_2_velocity, part_3_velocity, part_1_mean_force, part_2_mean_force, part_3_mean_force = divide_chart(all_velocity_list, velocity_1, velocity_2)
```

```python
def calculate_power(all_mean_force_list, all_velocity_list):
    power_list = []
    for i, value in enumerate(all_mean_force_list):
        power = all_mean_force_list[i] * all_velocity_list[i]
        power_list.append(power)
    return power_list
    
power_list = calculate_power(all_mean_force_list, all_velocity_list)
    
plt.figure(figsize=(5,5))
plt.title('Moc od prędkości')
plt.plot(all_velocity_list, power_list, 'm-o')
plt.xlabel('prędkość [m/s]')
plt.ylabel('moc [W]')
plt.grid()
plt.show()
```

### Drugi sposób liczenia prędkości
tablicę prędkości liczę z pochodnej przemieszczenia

```python
def calculate_velocity2(displacement_rebound_list, displacement_compression_list, sampling_freq):
    compression_velocity_list = []
    rebound_velocity_list = []
    for disp_array in displacement_rebound_list:
        velocity_rebound = np.mean(np.diff(disp_array))/(1/sampling_freq)
        velocity_rebound = velocity_rebound/1000
        rebound_velocity_list.append(velocity_rebound)
    for disp_array in displacement_compression_list:
        velocity_compression = np.mean(np.diff(disp_array))/(1/sampling_freq)
        velocity_compression = velocity_compression/1000
        compression_velocity_list.append(velocity_compression)
    return rebound_velocity_list, compression_velocity_list

rebound_velocity_list2, compression_velocity_list2 = calculate_velocity2(displacement_rebound_list, 
                                                                       displacement_compression_list,
                                                                       sampling_freq)
```
